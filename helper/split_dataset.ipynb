```python
import os
import shutil
import random

# Set the seed for reproducibility
seed = 1234
random.seed(seed)

# Define the main root folders and their corresponding train, test, and validation folders
main_root_folders = ['carakan', 'sandhangan', 'pasangan']
train_root = 'train'
test_root = 'test'
val_root = 'val'

# Define the percentages for train, test, and validation splits
train_percentage = 0.6
test_percentage = 0.2
val_percentage = 0.2

# Create the train, test, and validation root folders
os.makedirs(train_root, exist_ok=True)
os.makedirs(test_root, exist_ok=True)
os.makedirs(val_root, exist_ok=True)

# Iterate through each main root folder
for main_folder in main_root_folders:
    main_folder_path = os.path.join(main_folder)

    # Create the corresponding train, test, and validation subfolders for the current main folder
    train_folder = os.path.join(train_root, main_folder)
    test_folder = os.path.join(test_root, main_folder)
    val_folder = os.path.join(val_root, main_folder)
    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(test_folder, exist_ok=True)
    os.makedirs(val_folder, exist_ok=True)

    # Iterate through each label subfolder within the current main folder
    for label_folder in os.listdir(main_folder_path):
        label_folder_path = os.path.join(main_folder_path, label_folder)

        # Skip if the item is not a directory
        if not os.path.isdir(label_folder_path):
            continue

        # Get the list of images within the current label subfolder
        images = os.listdir(label_folder_path)

        # Shuffle the images
        random.shuffle(images)

        # Calculate the number of images for each split
        num_images = len(images)
        num_train = int(num_images * train_percentage)
        num_test = int(num_images * test_percentage)
        num_val = num_images - num_train - num_test

        # Split the images into train, test, and validation sets
        train_images = images[:num_train]
        test_images = images[num_train:num_train + num_test]
        val_images = images[num_train + num_test:]

        # Move the images to the respective train, test, and validation subfolders
        for image in train_images:
            src = os.path.join(label_folder_path, image)
            dst = os.path.join(train_folder, label_folder, image)
            os.makedirs(os.path.dirname(dst), exist_ok=True)
            shutil.copy(src, dst)

        for image in test_images:
            src = os.path.join(label_folder_path, image)
            dst = os.path.join(test_folder, label_folder, image)
            os.makedirs(os.path.dirname(dst), exist_ok=True)
            shutil.copy(src, dst)

        for image in val_images:
            src = os.path.join(label_folder_path, image)
            dst = os.path.join(val_folder, label_folder, image)
            os.makedirs(os.path.dirname(dst), exist_ok=True)
            shutil.copy(src, dst)

print("Dataset split completed!")
```

```python
import os

root_folders = ['train', 'test', 'val']
main_root_folders = ['carakan', 'pasangan', 'sandhangan']

# Function to count the number of images in each label folder
def count_images(folder):
    label_counts = {}
    for label_folder in os.listdir(folder):
        label_folder_path = os.path.join(folder, label_folder)
        if os.path.isdir(label_folder_path):
            count = len(os.listdir(label_folder_path))
            label_counts[label_folder] = count
    return label_counts

# Iterate through the root folders (train, test, val)
for root_folder in root_folders:
    print(f"{root_folder}:")
    for main_folder in main_root_folders:
        main_folder_path = os.path.join(root_folder, main_folder)
        if os.path.exists(main_folder_path):
            label_counts = count_images(main_folder_path)
            for label, count in label_counts.items():
                print(f"{main_folder}/{label}: {count} images")
    print()
```
