```python
import os
import cv2
from imgaug import augmenters as iaa
from imgaug.augmenters import contrast as iaa_contrast
import numpy as np


def augment_image(image):
    # Define the augmentation sequence
    seq = iaa.Sequential([
        iaa.Add((-10, 10), per_channel=0.5),  # Add random values to each pixel channel
        iaa.Multiply((0.9, 1.1), per_channel=0.5),  # Multiply each pixel channel by a random value
        iaa_contrast.LinearContrast((0.8, 1.2), per_channel=0.5),  # Adjust contrast of each pixel channel
        iaa.AdditiveGaussianNoise(loc=0, scale=(0.0, 0.05 * 255), per_channel=0.5),  # Add Gaussian noise to each pixel channel
    ])

    # Augment the image
    augmented_image = seq.augment_image(image)
    return augmented_image


def augment_images(input_folder, num_images):
    # Iterate over the subfolders in the input folder
    for root, dirs, files in os.walk(input_folder):
        for dir in dirs:
            subfolder_path = os.path.join(root, dir)

            # Iterate over the images in the subfolder
            for filename in os.listdir(subfolder_path):
                if filename.endswith((".jpg", ".png")):
                    # Load the image
                    input_path = os.path.join(subfolder_path, filename)
                    image = cv2.imread(input_path)
                    if image is None:
                        continue

                    # Generate augmented images
                    for i in range(num_images):
                        augmented_image = augment_image(image)

                        # Preserve white background
                        mask = np.all(augmented_image == [0, 0, 0], axis=-1)
                        augmented_image[mask] = [255, 255, 255]

                        # Save the augmented image
                        output_filename = f"{os.path.splitext(filename)[0]}_{i}.jpg"
                        output_path = os.path.join(subfolder_path, output_filename)
                        cv2.imwrite(output_path, augmented_image)


# Example usage
input_folder = "sandhangan"
num_images = 60

augment_images(input_folder, num_images)
```
