```python
#prepare
import os
import cv2 as cv
import numpy as np
import sys
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#OPENCV SETS
def list_dir(parent_dir):
    childs_dir = []
    for dir_name in os.scandir(parent_dir):
        if dir_name.is_dir():
            childs_dir.append(dir_name.name)
    return childs_dir
def cv_list_image(parent_dir):
    img_extensions = ['.jpg']
    img_lists = []
    iter = 1
    for file_name in os.listdir(parent_dir):
        if any(file_name.lower().endswith(ext) for ext in img_extensions ):
            file_route = str(parent_dir+'/'+file_name)
            print(file_route)
            img = cv.imread(file_route)
            newimg = cv_image_detect_101_new(image = img, offset=50)
            # cv_imshow(title=file_route,image = newimg, size = 4)
            # break
            cv_save(route = file_route, image = newimg, width = 200, height = 200)


def cv_imshow(title = "Image", image = None, size = 10):
    if image is None:
        print("[IMSHOW]Error: No image provided.")
        return
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

def cv_save(route="output.jpg", image=None, width=100, height=100):
    if image is None:
        print("Error: No image provided.")
        return

    if image.size == 0:
        print("Error: Empty image provided.")
        return

    if len(image.shape) < 2:
        print("Error: Invalid image provided.")
        return
    
    resized_image = cv.resize(image, (width, height))
    cv.imwrite(route, resized_image)


def cv_image_detect_101_new(image=None, offset = 0):
    if image is None:
        print("[IMAGE DETECT]Error: No image provided.")
        return None

    img_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    _, binary_img = cv.threshold(img_gray, 200, 255, cv.THRESH_BINARY_INV)

    contours, _ = cv.findContours(binary_img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    if len(contours) == 0:
        print("[IMAGE DETECT]Error: No contours found.")
        return None

    lx1, ly1, lx2, ly2 = [], [], [], []
    for contour in contours:
        x, y, w, h = cv.boundingRect(contour)
        lx1.append(x)
        ly1.append(y)
        lx2.append(x + w)
        ly2.append(y + h)

    x1 = min(lx1)
    y1 = min(ly1)
    x2 = max(lx2)
    y2 = max(ly2)
    width = x2 - x1
    height = y2 - y1
    Xc = x1 + width // 2
    Yc = y1 + height // 2

    wide_c = max(width, height)
    wide_c = int(wide_c + 2 * offset)
    wide_c = wide_c // 2
    x1 = Xc - wide_c
    y1 = Yc - wide_c
    x2 = Xc + wide_c
    y2 = Yc + wide_c

    if x1 < 0 : 
      x1 = 0
    if y1 < 0 : 
      y1 = 0
    if x2 > image.shape[1] : 
      x2 = image.shape[1]
    if y2 > image.shape[0] : 
      y2 = image.shape[0]

    cropped_img = image[y1:y2, x1:x2]
    return cropped_img

```
